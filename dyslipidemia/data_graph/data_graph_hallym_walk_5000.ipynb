# 데이터 불러오기 & 필요한 데이터 Filter

# 운동 활동량(TOTAL_STEP)에 따른 혈중 지질 농도(HDL, LDL, TC, TG, Waist)

```python
#8주간 하루 걸음을 실시한 집단의 혈중 지질 차이 pd 생성
walk_pd = {
    'Sample_ID' : [],
    'HDL' : [],
    'TG' : [],
    'LDL' : [],
    'Bun' : [],
    'Waist' : [],
    'TOTAL_STEP' : [],
    'MORNING_STEP' : [],
    'DAYTIME_STEP' : [],
    'EVENING_STEP' : [],
    'NIGHT_STEP' : []
}

walk_pd = pd.DataFrame(walk_pd)
```

```python
walk_pd.info()
```

```python
#walk_total_pd 생성
for index, row in walk_df.iterrows():

    HDL=int(row[12])-int(row[6])
    TG=int(row[11])-int(row[5])
    LDL=int(row[13])-int(row[7])
    Bun=int(row[10])-int(row[4])
    Waist=int(row[9])-int(row[3])
    total_step=int(row[14])
    morning_step=int(row[16])
    daytime_step=int(row[17])
    evening_step=int(row[18])
    night_step=int(row[15])
    
    #데이터 추가
    walk_pd.loc[len(walk_pd)] = [row[0],HDL,TG,LDL,Bun,Waist,total_step,morning_step,daytime_step,evening_step,night_step]
    
walk_pd
```

```python
#walk_pd의 혈중 지질 농도 평균 변화
walk_pd_tg_mean = walk_pd['TG'].mean()
walk_pd_ldl_mean = walk_pd['LDL'].mean()
walk_pd_hdl_mean = walk_pd['HDL'].mean()
walk_pd_bun_mean = walk_pd['Bun'].mean()
walk_pd_waist_mean = walk_pd['Waist'].mean()


print(f'8주 후 TG 평균 감소: {walk_pd_tg_mean}')
print(f'8주 후 LDL 평균 감소: {walk_pd_ldl_mean}')
print(f'8주 후 HDL 평균 감소: {walk_pd_hdl_mean}')
print(f'8주 후 BUN 평균 감소: {walk_pd_bun_mean}')
print(f'8주 후 Waist 평균 감소: {walk_pd_waist_mean}')
```

# 논문 내용 % 감소 비율 실제 확인

```python
#1차 방문 (8주간 운동 전 평균)
walk_df_tg_1mean = walk_df['TG_1'].mean()
walk_df_ldl_1mean = walk_df['LDL_1'].mean()
walk_df_hdl_1mean = walk_df['HDL_1'].mean()
walk_df_bun_1mean = walk_df['BUN_1'].mean()
walk_df_waist_1mean = walk_df['Waist_1'].mean()

print(f'8주 전 TG 평균: {walk_df_tg_1mean}')
print(f'8주 전 LDL 평균: {walk_df_ldl_1mean}')
print(f'8주 전 HDL 평균: {walk_df_hdl_1mean}')
print(f'8주 전 BUN 평균: {walk_df_bun_1mean}')
print(f'8주 전 Waist 평균: {walk_df_waist_1mean}')
```

```python
#2차 방문 (8주간 운동 후 평균)
walk_df_tg_2mean = walk_df['TG_2'].mean()
walk_df_ldl_2mean = walk_df['LDL_2'].mean()
walk_df_hdl_2mean = walk_df['HDL_2'].mean()
walk_df_bun_2mean = walk_df['BUN_2'].mean()
walk_df_waist_2mean = walk_df['Waist_2'].mean()

print(f'8주  후 TG 평균: {walk_df_tg_2mean}')
print(f'8주 후 LDL 평균: {walk_df_ldl_2mean}')
print(f'8주 후 HDL 평균: {walk_df_hdl_2mean}')
print(f'8주 후 BUN 평균: {walk_df_bun_2mean}')
print(f'8주 후 Waist 평균: {walk_df_waist_2mean}')
```

```python
# 퍼센트 비율 감소 = (증가량 또는 감소량 / 시작 값) × 100
# 증가량 또는 감소량 = |나중 값 – 시작 값|

walk_tg = (abs(walk_df_tg_2mean-walk_df_tg_1mean)/walk_df_tg_1mean) * 100
walk_ldl = (abs(walk_df_ldl_2mean-walk_df_ldl_1mean)/walk_df_ldl_1mean) * 100
walk_hdl = (abs(walk_df_hdl_2mean-walk_df_hdl_1mean)/walk_df_hdl_1mean) * 100
walk_bun = (abs(walk_df_bun_2mean-walk_df_bun_1mean)/walk_df_bun_1mean) * 100
walk_waist = (abs(walk_df_waist_2mean-walk_df_waist_1mean)/walk_df_waist_1mean) * 100

print(f'8주간 TG 평균 감소 비율: {walk_tg}')
print(f'8주간 LDL 평균 감소 비율: {walk_ldl}')
print(f'8주간 HDL 평균 감소 비율: {walk_hdl}')
print(f'8주간 BUN 평균 감소 비율: {walk_bun}')
print(f'8주간 Waist 평균 감소 비율: {walk_waist}')
```
