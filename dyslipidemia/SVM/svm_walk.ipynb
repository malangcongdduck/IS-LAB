```python
from sklearn import svm
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Hdl, bmi, tg, waist, bun, 에너지 소비량 이 활동량에 영향을 주는 인자로 작용 (에너지 소비량 없음)
df = pd.read_excel('./활동량.xlsx') 
print(df)
```

```python
#보조 호르몬 요법을 받고 있는 Cohort 3 제거 Filter 적용
df = df[(df['COHORT'] != 3)]
```

```python
df.isnull().sum()
```

```python
#하루 에너지 소비량 걸음 수 (Y) = 0.0(보폭) + 0.04 × (X)
#BMI 데이터 너무 적어서 제외하고 보기
df=df[['HDL_1','HDL_2','BUN_1','BUN_2','Waist_1','Waist_2','TOTAL_STEP']]
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df
```

```python
x_no_HDL=[]
y_HDL=[]

x_no_BUN=[]
y_BUN=[]

x_no_Waist=[]
y_Waist=[]
#x,y 배열 생성
for index, row in df.iterrows():

    HDL=int(row[0])-int(row[1])
    BUN=int(row[2])-int(row[3])
    Waist=int(row[4])-int(row[5])
    STEP=int(row[6])
    
    #데이터 추가
    x_no_HDL.append([BUN, Waist, STEP])
    y_HDL.append(HDL)
    
    x_no_BUN.append([HDL, Waist, STEP])
    y_BUN.append(BUN)
    
    x_no_Waist.append([HDL, BUN, STEP])
    y_Waist.append(Waist)
```

```python
np.random.seed(0)
x_train,x_test,y_train,y_test=train_test_split(x_no_HDL,y_HDL,train_size=0.8)
#x_train,x_test,y_train,y_test=train_test_split(x_no_BUN,y_BUN,train_size=0.8)
#x_train,x_test,y_train,y_test=train_test_split(x_no_Waist,y_Waist,train_size=0.8)
```

```python
s=svm.SVC(gamma=0.0015, C=5)
s.fit(x_train,y_train)
```

```python
res=s.predict(x_no_HDL)
correct=[i for i in range(len(res)) if res[i]==y_HDL[i]]
accuracy=len(correct)/len(res)
print("정확률=",accuracy*100,"%")
```
