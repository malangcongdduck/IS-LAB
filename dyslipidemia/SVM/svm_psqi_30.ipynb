```python
from sklearn import svm
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('./최종 데이터 그래프그리기용.xlsx') 
print(df)
```

```python
psqi_df=df[['Sample_ID','GROUP','COHORT','SEX','AGE','TG_1','TG_2','PSQI_TOTAL_1','PSQI_TOTAL_2']]
```

```python
psqi_df
```

```python
#보조 호르몬 요법을 받고 있는 Cohort 3 제거 Filter 적용
psqi_df = psqi_df[(psqi_df['COHORT'] != 3)]
psqi_df
```

```python
psqi_df=psqi_df.dropna()
psqi_df.reset_index(drop=True, inplace=True)
psqi_df.isnull().sum()
```

```python
#x,y 배열 생성
x=[]
y=[]

for index, row in psqi_df.iterrows():

    psqi_1=int(row[7])
    psqi_2=int(row[8])
    tg=int(row[5])-int(row[6])

    
    #데이터 추가
    x.append([psqi_1, psqi_2])
    y.append(tg)
```

```python
len(x), len(y)
```

```python
np.random.seed(0)
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
s=svm.SVC(gamma=50, C=1)
s.fit(x_train,y_train)
```

```python
res=s.predict(x_train)
correct=[i for i in range(len(res)) if res[i]==y_train[i]]
accuracy=len(correct)/len(res)
print("정확률=",accuracy*100,"%")
```

```python
y_pred = s.predict(x_test) # 테스트
print("예측된 라벨:", y_pred)
print("ground-truth 라벨:", y_test)

print()
print("prediction accuracy: {:.2f} %".format(np.mean(y_pred == y_test)*100)) # 예측 정확도
```
